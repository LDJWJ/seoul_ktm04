## 조건 판단 - if
```
if  (   ):
____실행문1

()의 값이 참이면 실행문1을 실행한다.
()의 값이 참이 아니면 실행문1을 실행하지 않는다.
```

```
if  (   ):
____실행문1
else:
____실행문2

()의 값이 참이면 실행문1을 실행한다.
()의 값이 참이 아니면 실행문2을 실행한다.
```

```
if   (A):
____실행문1
elif (B):
____실행문2
else:
____실행문3

(A)의 값이 참이면 실행문1을 실행한다.
(A)가 아니고, (B)의 값이 참이면 실행문2을 실행한다.
(A)도 아니고, (B)도 아니면 실행문3을 실행한다.
```

````python
emo = ['행복', '감사', '평화', '맑음']
neg = ['슬픔', '화남', '우울', '월요일']
````

````python
mem = input("당신의 현재 상태는?")

if  (mem in emo):
  print("현재 긍정적인 상태입니다.")
  print("당신의 감정 단어는", mem)
elif (mem in neg):
  print("현재 부정적인 상태입니다.")
  print("당신의 감정 단어는", mem)
else:
  print("오늘은 평범한 날이네요.")
  print("당신의 단어는", mem)  
````

### 라이브러리 불러오기 
```
import  프로그램이름
import  프로그램이름  as 약자

A ( AB AC AD)
AB에 접근할 때, A.AB를 불러올때,
import A.AB

우리는 AB만 불러오고싶다. 
import A  => A ( AB AC AD)
from A import AB    # A중에 AB라는 친구(기능)만 가져오겠다.
from A import AB, AC  # A중에 AB, AC라는 두친구(기능)을 가져오겠다.
```
